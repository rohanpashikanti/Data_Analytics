# Binning
* Binning: Grouping of values into bins"
* Converts numeric into categorical variables
* Group a set of numerical values into a set of "bins"
* "price" is an feature range from 5,000 to 45,500



```python
ab.describe()
```

```python
bins = np.linspace(min(ab['weight']),max(ab['weight']),4)
GroupNames = ['Low','Medium','High'] #Added a label to match the number of bins
ab['weight'] = pd.cut(ab['weight'], bins, labels=GroupNames, include_lowest=True)
```

```python

import matplotlib.pyplot as plt
plt.hist(ab["weight"])
```
